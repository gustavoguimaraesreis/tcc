```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 

# Tres meses antes do ocorrido:
sns.set_theme(style="whitegrid")
tweets = pd.read_json("tweets/facebook/tres-meses/facebook data leak_2021-01-02_2021-04-02.json",
                     lines=True, nrows=94, convert_dates=False)


tweets["Date"] = pd.to_datetime(tweets.Date, format="%d/%m/%Y")
tweets["Month"] = tweets["Date"].dt.month


f, ax = plt.subplots(figsize=(15, 10))

sns.countplot(x= tweets["Month"])
# sns.histplot(x=tweets["Month"])
ax.set_xlabel("Meses", size = 12 ) 
ax.set_ylabel("Quantidade de Tweets" , size = 12 )
ax.set_xticklabels(["Janeiro", "Fevereiro", "Mar√ßo", "Abril"])
ax.set_xticks(minor=False)
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+0.05, p.get_height()+20), fontsize = 12) 


# chart = sns.displot(x= tweets["Month"], kind="kde")
# chart.set_xlabels("Meses")
# chart.set_ylabels("Densidade")
```
